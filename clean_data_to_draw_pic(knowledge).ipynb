`Frequency`（`频数`） ：知识点`K_i`在多少题目中出现过

`Proportion of practices`（`频率`）：有知识点 `i` 出现的题目的数量占总题目数量的百分比（$\frac{频数}{题目总数}$）
> 若一道题只考察一个知识点，那所有知识点在题目中出现的频率相加应该为`1`，但是编程题目中一道题往往涉及多个知识点，所以所有知识点在题目中出现的频率相加大于 `1`.

`Cumulative proportion of practices`（`累计频率`）：用来累加知识点 `K_i` 以上所有知识点在题目中的出现频率。用来观察增长的速率。
> $k3_{Cumulative proportion of practices} = k2_{Cumulative proportion of practices}+k3_{Proportion of practices}$

`Proportion of practices（%）`（`频率（%）`）：$ Proportion of practices \times 100\%$

`Cumulative proportion of practices(%)`(`累计频率（%）`)：本来与累计频率一样，是对频率的累加，但是累加完 $\times 100\%$ 发现结果超 `300\%`，所以这里做了规一化处理，$\frac{Ki_{Cumulative proportion of practices}}{K176_{Cumulative proportion of practices}}$


```python
practices_num = int(632)
Cumulative_proportion_of_practices = float(30.356013)

def calculate(Cumulative):
    return Cumulative/Cumulative_proportion_of_practices*100

##################################### 英文字段名 #####################################
# temp['Proportion of practices'] = temp / practices_num # 计算所出现题目占所有题目的比例

# temp['Cumulative proportion of practices'] = temp['Proportion of practices'].cumsum()  # 计算累计频率
# temp['Proportion of practices(%)'] = temp['Proportion of practices'].apply(lambda x: "%.2f%%" % (x*100))  # 以百分比显示频率
# temp['Cumulative proportion of practices(%)'] = temp['Cumulative proportion of practices'].apply(lambda x: "%.2f%%" % (calculate(x)))  # 以百分比显示累计频率，归一化累加
# # 可视化显示
# temp.style.bar(subset=['Proportion of practices','Cumulative proportion of practices'], color='green',width=100)

##################################### 中文字段名 #####################################
temp['频率'] = temp / practices_num # 计算所出现题目占所有题目的比例

temp['累计频率'] = temp['频率'].cumsum()  # 计算累计频率
temp['频率(%)'] = temp['频率'].apply(lambda x: "%.2f%%" % (x*100))  # 以百分比显示频率
temp['累计频率(%)'] = temp['累计频率'].apply(lambda x: "%.2f%%" % (calculate(x)))  # 以百分比显示累计频率，归一化累加
# 可视化显示
#temp.style.bar(subset=['频率','累计频率'], color='green',width=100)
df_table = temp.style.bar(subset=['频率','累计频率'], color='green',width=100)
df_table
```
