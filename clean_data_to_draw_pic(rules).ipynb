## 查询学习者一共做了多少习题

用到的数据：`/Users/eleanor/学习路径实验/rulesAndError_to_knowledge/data/compile_errors/compile_err_res_before_merge.txt`

数据字段含义：

`repo_path`, `shixun_id`, `step_id`, `error_id_list`

```python
with open('/Users/eleanor/学习路径实验/rulesAndError_to_knowledge/data/compile_errors/compile_err_res_before_merge.txt', 'r') as f:
    learner_error = f.readlines()  #txt中所有字符串读入data

    i = 0
    shixun_step = []
    for line in learner_error:
        #if i < 100:
            line_move_t = line.split('\t')
            line_shixun_step = []
            # shixun_id 和 step_id 写入 list： line_shixun_step
            line_shixun_step.append(line_move_t[1])
            line_shixun_step.append(line_move_t[2])
            shixun_step.append(line_shixun_step)
            #print(i)
        #i = i + 1
    #print(shixun_step)
    #print(len(shixun_step))
```

```python
# 去除嵌套 List 中的重复项

shixun_step_uniq = [list(t) for t in set(tuple(_) for _ in shixun_step)]
print(shixun_step_uniq)
print(len(shixun_step_uniq))
```

```python
# 比较嵌套 List 中的元素，即先比较子 List 中的第一个元素（shixun_id），再比较子 List 中的第二元素（step_id），按大小排序
from functools import cmp_to_key

def comparable_fun(a, b):
    if a[0]>b[0]:
        return 1
    elif a[0]==b[0]:
        return a[1]>b[1]
    else:
        return -1

shixun_step_uniq.sort(key=cmp_to_key(comparable_fun))
print(shixun_step_uniq)
print(len(shixun_step_uniq))
```

```python
# 计算激活实训个数
shixun_active = []
for i in shixun_step_uniq:
    shixun_active.append(i[0])
print(shixun_active)
```

```python
# 去除重复项
shixun_active_uniq = list(set(shixun_active))
print(shixun_active_uniq)
print(len(shixun_active_uniq))
```

## 统计 Sonar 的 Rules 种类数量

用到的数据：`/Users/eleanor/学习路径实验/rules_match/data/sonar_rule.csv`

```python
import pandas as pd

# read csv
rules = pd.read_csv('/Users/eleanor/学习路径实验/rules_match/data/sonar_rule.csv')
print(rules.info())
print("*************")
print(rules.head(5))
```

```python
# 提出 csv 中的列 type 和 defaultSeverity
rules_type = []
rules_defaultSeverity = []
for index, row in rules.iterrows():
    i = row['type']
    j = row['defaultSeverity']
    rules_type.append(i)
    rules_defaultSeverity.append(j)
print(rules_type) 
```

```python
# 统计 csv 列 type 中每类元素出现的次数
# 统计 csv 列 defaultSeverity 中每类元素出现的次数
from collections import Counter

count_rules_type = Counter(rules_type)
print(count_rules_type)

count_rules_defaultSeverity = Counter(rules_defaultSeverity)
print(count_rules_defaultSeverity)
```
